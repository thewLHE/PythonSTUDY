```python
# 이에서 점수를 메기는 것을 역으로 실행
seqA="ATGTCGACTA"
seqB="AAATCGCGCGTGCTG"

m,n=len(seqA),len(seqB)

s0=[0 for i in range(n)]
Sij=[]
for i in range(m):
    Sij.append(s0.copy())
#Initialize
for i in range(n):
    Sij[0][i]=-3*i
for j in range(m):
    Sij[j][0]=-3*j

def w(a,b):
    if(a==b and a!="-"):
        return 8
    elif(a=="-" or b=="-"):
        return -3
    else:
        return -5

for i in range(m-1):
    for j in range(n-1):
        #Sij[i+1][j+1]=max(w(seqA[i+1],seqB[j+1]),w(seqA[i+1],'-'),w('-',seqB[j+1]))
        Sij[i+1][j+1]=max( Sij[i][j]+w(seqA[i],seqB[j]),\
                          Sij[i][j+1]+w(seqA[i],"-"),\
                          Sij[i+1][j]+w('-',seqB[j]) )
        
for i in range(m):
    for j in range(n):
        print("{:3d}".format(Sij[i][j]),end=" ")
    print("\n")
```

```python
#solv
seqA="ATGTCGACTA"
seqB="AAATCGCGCGTGCTG"

n,m=len(seqA),len(seqB)

s0=[0 for i in range(m+1)]
S=[]


for i in range(n+1):
    S.append(s0.copy())#score matrix initialize
#Initialize
for i in range(n+1):
    S[i][0]=-3*i
for j in range(m+1):
    S[0][j]=-3*j
    
def w(a,b):
    if(a==b):
        return 8
    else:
        return -5
gap=-3

for i in range(n):#calcul score value
    for j in range(m):
            S[i+1][j+1]=max(S[i][j] + w(seqA[i],seqB[j]),\
                          S[i][j+1]+gap,\
                          S[i+1][j]+gap )
            
for d in '-'+seqB:
    print("{:>4s}".format(d),end="")
print("\n")
for i in range(1,n+1):
    print("{:^4s}".format(seqA[i-1]),end="")
    
    for j in range(1,m+1):
        print('{:4d}'.format(S[i][j]),end='')
    print("\n")
        
#
AlignmentA=""
AlignmentB=""
i=len(seqA)
j=len(seqB)

while (i > 0 or j > 0):#print ali
    if (i > 0 and j > 0 and S[i][j] == S[i][j] +w(seqA[i-1],seqB[j-1])):
        AlignmentA = seqA[i] + AlignmentA
        AlignmentB = seqB[j] + AlignmentB
        i = i - 1
        j = j - 1
    
    elif (i > 0 and S[i][j] == S[i-1][j] +gap):
        AlignmentA = seqA[i-1] + AlignmentA
        AlignmentB = "−" + AlignmentB
        i = i - 1
    
    else:
        AlignmentA = "−" + AlignmentA
        AlignmentB = seqB[j-1] + AlignmentB
        j = j - 1
#
print(AlignmentA)
print(AlignmentB)
```
