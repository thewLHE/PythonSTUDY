```python
#solv
seqA="ATGTCGACTA"
seqB="AAATCGCGCGTGCTG"

m,n=len(seqA),len(seqB)

s0=[0 for i in range(n)]
Sij=[]
for i in range(m):
    Sij.append(s0.copy())
#Initialize
for i in range(n):
    Sij[0][i]=-3*i
for j in range(m):
    Sij[j][0]=-3*j

def w(a,b):
    if(a==b and a!="-"):
        return 8
    elif(a=="-" or b=="-"):
        return -3
    else:
        return -5

for i in range(m-1):
    for j in range(n-1):
        print(seqA[i],seqB[j],w(seqA[i],seqB[j]),w(seqA[i],'-'),w('-',seqB[j]))
        #Sij[i+1][j+1]=max(w(seqA[i+1],seqB[j+1]),w(seqA[i+1],'-'),w('-',seqB[j+1]))
        Sij[i+1][j+1]=max( Sij[i][j]+w(seqA[i],seqB[j]),\
                          Sij[i][j-1]+w(seqA[i],"-"),\
                          Sij[i-1][j]+w('-',seqB[j]) )
        
print(Sij)
```
